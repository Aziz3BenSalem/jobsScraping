```python
from bs4 import BeautifulSoup
import requests
import csv
from itertools import zip_longest

page_num = 0

while True:
    r = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb&q=&start={page_num}")
    print(r)

    src = r.content
    soup = BeautifulSoup(src, 'lxml')
    job_title = []
    company_names = []
    city = []
    skill = []

    jobs_titles = soup.find_all("h2", {"class": "css-m604qf"})
    companies_titles = soup.find_all("a", {"class": "css-17s97q8"})
    cities = soup.find_all("span", {"class": "css-5wys0k"})
    skills = soup.find_all("div", {'class': 'css-y4udm8'})

    for i in range(len(jobs_titles)):
        job_title.append(jobs_titles[i].text)
        company_names.append(companies_titles[i].text)
        city.append(cities[i].text)
        skill.append(skills[i].text)

    my_list = [job_title, company_names, city, skill]
    exported = zip_longest(*my_list)

    with open("C:\\scraping\\jobs.csv", "a", newline='', encoding='utf-8') as myfile:
        wr = csv.writer(myfile)
        if page_num == 0:
            wr.writerow(["job_title", "company_names", "city", "skill"])
        wr.writerows(exported)

    page_num += 1
    print(page_num)

    if page_num > 673:
        print("pages ended")
        break

```

```python

input_file_path = "C:\\scraping\\jobs.csv"

with open(input_file_path, "r", encoding='utf-8') as infile:
    reader = csv.reader(infile)
    data = [row for row in reader]

non_empty_data = [row for row in data if any(cell.strip() for cell in row)]

with open(input_file_path, "w", newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(non_empty_data)

print("Empty rows removed.")
```
